```python
import numpy as np

bow2_file = 'bow2.words'
bow5_file = 'bow5.words'
dependency_file = 'deps.words'

#read the files
with open(bow2_file) as f:
    bow2 = f.readlines()
    
with open(bow5_file) as f:
    bow5 = f.readlines()
    
    
    
word2 = []
word5 = []
word_deps = []

vector2 = dict()
vector5 = dict()
vector_deps = dict()
    

#bow2
for l in bow2:
    #split line
    split_line = str.split(l)
    
    #first one is the word - string
    word = split_line[0]
    word2.append(word)
    
    #remaining elements are the embedding values - float
    vector = []
    for vd in split_line[1:]:
        vector.append(float(vd))
        
    vector2[word] = vector
    
    
#bow5
for l in bow5:
    #split line
    split_line = str.split(l)
    
    #first one is the word - string
    word = split_line[0]
    word5.append(word)
    
    #remaining elements are the embedding values - float
    vector = []
    for vd in split_line[1:]:
        vector.append(float(vd))
        
    vector5[word] = vector
```
